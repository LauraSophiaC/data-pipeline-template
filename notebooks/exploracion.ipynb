```python
import pandas as pd

df = pd.read_parquet("../data/medallion/gold/dataset_training_20250723.parquet")
df.head(100)

```

```python
df.to_csv("../data/medallion/gold/dataset_training_20250723.csv", index=False)
```

```python
print("--porcentaje de clicks vs no clicks:")
print(df["clicked"].value_counts(normalize=True))

print("--Clicks por value_prop:")
resumen = df.groupby("value_prop")["clicked"].agg(["sum", "count"])
print(resumen.rename(columns={"sum": "clicks", "count": "total"}))
print("--Value_prop que genera más pagos:")
print(df.groupby("value_prop")["past_pays_total"].sum().sort_values(ascending=False))
print("--Value_prop con más interacción:")
print(df["value_prop"].value_counts())
print("--Análisis por usuario:")
print(df.groupby("user_id")[["clicked", "past_taps", "past_pays_total"]].mean().head())


```
